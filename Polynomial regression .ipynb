```python
##import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
X=6*np.random.rand(100,1)-3
y=0.5 * X**2 +1.5*X+2+np.random.randn(100,1)
plt.scatter(X,y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
####lets implement a simple linear regression 
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
```

```python
regression_1.fit(X_train,y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,regression_1.predict(X_test))
```

```python
print(score)
```

```python
### Lets visualize this model 
plt.plot(X_train,regression_1.predict(X_train),color='r')
plt.scatter(X_train,y_train)
```

```python
###### here for this data set linear regression model fails because it not best fit line so that our accuracy score is very lesss 
###to improve accracy we are using polymomial regression model 
#### lets apply polymial transformation 
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
```

```python
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,y_train)
y_pred=regression.predict(X_test_poly)
score=r2_score(y_test,y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```
